```python
import trectools as tr
import pandas as pd
```

```python
mqdf = pd.read_csv('./MQ2008-agg/agg.txt', sep='\s+', header=None) #read mq2008 letor 4 file
mqdf.head(5)
```

```python
# transform mq 2008
def ranking_value(x):
    if x == 'NULL':
        return 0
    else:
        return x
    
mqdf['relevance'] = mqdf[0]
mqdf['qid'] = mqdf[1].apply(lambda x: x.split(':')[1])
for i in range(2, 27):
    mqdf['ranking_'+str(i-1)] = mqdf[i].apply(lambda x: ranking_value(x.split(':')[1]))

mqdf['docid'] = mqdf[29]
ranking_columns = ["ranking_{i}" for i in range(1,26)]
ranking_mqdf = mqdf[['qid', 'docid', 'relevance'] + ranking_columns]
ranking_mqdf.head(5)
```

```python
ranking_mqdf.to_csv('./MQ2008-agg/transformed_agg.csv', sep="\t" ,index=False)
```
